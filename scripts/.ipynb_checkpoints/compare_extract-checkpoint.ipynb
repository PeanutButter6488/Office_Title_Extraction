```python
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
f = open('../test_files/test3.txt', 'r')
titles = f.readlines()
f.close()
```

```python
f = open('../test_files/test5.txt', 'w')
for i in range(len(titles)):
    index = titles[i].index(":")
    f.write(titles[i][:index] + '\n')
f.close()
```

```python
address = pd.read_csv('../named_entities/cbdb_entity_addresses.csv')
```

```python
address.sample(3)
```

```python
offices = pd.read_csv('../named_entities/cbdb_entity_offices.csv')
```

```python
offices.sample(3)
```

```python
f = open('../test_files/test5.txt', 'r')
target = f.readlines()
title = {}
for i in range(len(target)):
    target[i] = target[i].strip()
title['Title'] = target
```

```python
df = pd.DataFrame(title)
```

```python
df['address_first2'] = 0
df['address_last2'] = 0
df['office_first2'] = 0
df['office_last2'] = 0
```

```python
for i in tqdm(range(len(target))):
    for j in df.index:
        if target[i][:2] == address['addr_chn'][j][:2]:
            df['address_first2'][i] = 1
            break
        if target[i][-2:] == address['addr_chn'][j][-2:]:
            df['address_last2'][i] = 1
            break
```

```python
df
```

```python
for i in tqdm(range(len(target))):
    for j in df.index:
        if target[i][:2] == offices['office_name'][j][:2]:
            df['office_first2'][i] = 1
            break
        if target[i][-2:] == offices['office_name'][j][-2:]:
            df['office_last2'][i] = 1
            break
```

```python
df
```

```python
count = 0
for i in df.index:
    if ~(df['address_first2'][i].all()) & ~(df['address_last2'][i].all()) & ~(df['office_first2'][i].all()) & ~(df['office_last2'].all()):
        count += 1
count
```

```python
df.to_csv('../test_files/compare_extract.csv')
```
