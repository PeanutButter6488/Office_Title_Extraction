```python
import numpy as np
import pandas as pd
import opencc
from tqdm import tqdm
```

```python
f = open('../test_files/test3.txt', 'r')
titles = f.readlines()
f.close()
```

```python
f = open('../test_files/test5.txt', 'w')
for i in range(len(titles)):
    index = titles[i].index(":")
    f.write(titles[i][:index] + '\n')
f.close()
```

```python
address = pd.read_csv('../named_entities/cbdb_entity_addresses.csv')
```

```python
address.sample(3)
```

```python
offices = pd.read_csv('../named_entities/cbdb_entity_offices.csv')
```

```python
offices.sample(3)
```

```python
f = open('../test_files/test5.txt', 'r')
target = f.readlines()
title = {}
for i in tqdm(range(len(target))):
    target[i] = target[i].strip()
    converter = opencc.OpenCC('s2t.json')
    target[i] = converter.convert(target[i])    # convert from simplified chinese to traditional chinese
title['Title'] = target
```

```python
title['Title'][0]
```

```python
df = pd.DataFrame(title)
```

```python
df['faddress_first2'] = 0       # first two words of the title match the first two words of sample address
df['laddress_first2'] = 0       # last two words of the title match the first two words of sample address
df['faddress_last2'] = 0         
df['laddress_last2'] = 0
df['foffice_first2'] = 0
df['loffice_first2'] = 0
df['foffice_last2'] = 0
df['loffice_last2'] = 0
```

```python
for i in tqdm(range(len(target))):
    for j in address.index:
        # tries to match the first two words of the title 
        if target[i][:2] == address['addr_chn'][j][:2]:      
            df['faddress_first2'][i] = 1
            break
        if target[i][-2:] == address['addr_chn'][j][:2]:
            df['laddress_first2'][i] = 1
            break
        if target[i][:2] == address['addr_chn'][j][-2:]:
            df['faddress_last2'][i] = 1
            break
        if target[i][-2:] == address['addr_chn'][j][-2:]:
            df['laddress_last2'][i] = 1
            break
```

```python
df.sample(3)
```

```python
for i in tqdm(range(len(target))):
    for j in offices.index:
        if target[i][:2] == offices['office_name'][j][:2]:
            df['foffice_first2'][i] = 1
            break
        if target[i][-2:] == offices['office_name'][j][:2]:
            df['loffice_first2'][i] = 1
            break
        if target[i][:2] == offices['office_name'][j][-2:]:
            df['foffice_last2'][i] = 1
            break
        if target[i][-2:] == offices['office_name'][j][-2:]:
            df['loffice_last2'][i] = 1
            break
```

```python
df.sample(3)
```

```python
df.to_csv('../test_files/compare_extract.csv')
```
