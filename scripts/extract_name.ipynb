从右面切人名：
    - 只考虑 l_office_first2 or l_office_last2 都为0的情况
    - 以前两个结果都为0，删掉后其中一个变为1的话
     ==> 有人名，前面的为官职

```python
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv('../test_files/compare_extract_nozero.csv')
```

```python
df
```

```python
df = df[ df['Title'].apply(lambda x: len(x) >= 5) ]
```

```python
df
```

```python
offices = pd.read_csv('../named_entities/cbdb_entity_offices.csv')
offices.sample(3)
```

```python
extracted = []
for i in tqdm(df.index):     
    for k in range(4):
        if (df['loffice_first2'][i] == 0) | (df['loffice_last2'][i] == 0):
            for j in offices.index:
                substring = df['Title'][i][:-k]
                if substring[-2:] == offices['office_name'][j][:2]:
                    extracted.append(substring + "\n")
                    break
                if substring[-2:] == offices['office_name'][j][-2:]:
                    extracted.append(substring + "\n")
                    break
            else:
                continue
            break
            
```

```python
extracted
```

```python
f = open("../test_files/titles_no_name.txt", "w")
for i in range(len(extracted)):
    f.write(extracted[i])
f.close()
    
```
